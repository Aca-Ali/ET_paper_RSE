```python
def convert_index_to_datetime(df):
    def parse_date_hour(date_str):
        # Split the string at the space to separate date and hour parts.
        date_part, hour_part = date_str.split(' ')
        # Ensure the hour is two digits (pad with a leading zero if necessary)
        hour_part = hour_part.zfill(2)
        # Recombine the parts
        new_str = f"{date_part} {hour_part}"
        return pd.to_datetime(new_str, format='%Y%m%d %H')
    # Apply the converter to the index
    df = df.copy()
    df.index = df.index.map(parse_date_hour)
    return df
```
